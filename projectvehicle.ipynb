### which types of vehicle grew most in the year between 2009-2015 ?

```python
#clean column names
data.columns= data.columns.str.replace(' ','')
data.index= data.index.str.strip()
```

```python
#step1 : making a list
vehicles_list_2009 = ['Buses_2009','Taxis_2009','Light Motor Vehicles(Passengers)_2009','Goods Carrier Vehicles(a)_2009', 
                    'Two-Wheeler_2009','Cars_2009', 'Jeeps_2009','Miscellaneous(b)_2009']
vehicles_list_2015 = ['Buses_2015','Taxis_2015','Light Motor Vehicles(Passengers)_2015','Goods Carrier Vehicles(a)_2015', 
                      'Two-Wheeler_2015','Cars_2015', 'Jeeps_2015','Miscellaneous(b)_2015']
```

```python
#step2:loop through both the list

growth ={}

for v09,v15 in zip(vehicles_list_2009,vehicles_list_2015):
    if v09 in data.columns and v15 in data.columns:
        total_2009 = data[v09].sum()
        total_2015 = data[v15].sum()
        absolute_growth = total_2015 - total_2009
        vehicle_name = v09.replace('_2009','')
        growth[vehicle_name] = absolute_growth
        
most_vehicle_growth = max(growth,key=growth.get)
print("vehicle_with_most_growth : " , most_vehicle_growth)
```

```python
#creating bar chart
vehicle_names =list(growth.keys())
vehicle_growth = list(growth.values())
```

```python
plt.figure(figsize=(5,4))
plt.bar(vehicle_names,vehicle_growth,color='skyblue')
plt.xlabel("vehicle type")
plt.ylabel("Increase in no. of vehicles")
plt.title("rise of vehicles 2009 vs 2015")
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.tight_layout()
plt.show()
```

### TIME SERIES FORECASTING

```python
maharashtra_data = data.loc['Maharashtra']
print(maharashtra_data)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
two_wheeler_data = {                   #dictionary is created
       2009:10212360.0,
       2010:11181762.0,
       2011:12429011.0,
       2012:13921763.0,
       2013:15457123.0,
       2014:16910395.0,
       2015:18603835.0
}
```

```python
df = pd.DataFrame(list(two_wheeler_data.items()),columns = ['years','two_wheeler_count'])   #converted into df
```

```python
X = df[['years']]
Y = df['two_wheeler_count']
```

```python
print(X.shape)
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
future = pd.DataFrame({'years':['2016','2017','2018','2019','2020']})
```

```python
predictions = model.predict(future)
```

```python
future['predicted_two_wheeler_count'] = predictions
print(future)
```

```python
pd.options.display.float_format = '{:,.0f}'.format
print(future)
```

```python
plt.figure(figsize=(5,4))
plt.bar(future.years,future.predicted_two_wheeler_count,color='skyblue')
plt.xlabel("pred_years")
plt.ylabel("pred_counts")
plt.title("pred")
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.tight_layout()
plt.show()
```
